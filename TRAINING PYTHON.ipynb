```python
#Spy Number

n = int(input())

sum = 0
product = 1

while n > 0:

    digit = n % 10

    sum += digit
    product *= digit

    n //= 10

if sum == product:
    print("Spy Number")
else:
    print("Not Spy Number")
```

```python
#neon number

n = int(input())

square = n * n

sum = 0

for i in str(square):
    sum += int(i)

if sum == n:
    print("Neon Number")
else:
    print("Not Neon Number")
```

```python
#Peterson Number

n = int(input())

temp = n
sum = 0

while temp > 0:

    digit = temp % 10

    fact = 1

    for i in range(1, digit + 1):
        fact *= i

    sum += fact

    temp //= 10

if sum == n:
    print("Peterson Number")
else:
    print("Not Peterson Number")
```

```python
#Twisted Prime

n = int(input())

rev = int(str(n)[::-1])

prime1 = True
prime2 = True

for i in range(2, n):

    if n % i == 0:
        prime1 = False

for i in range(2, rev):

    if rev % i == 0:
        prime2 = False

if prime1 and prime2:
    print("Twisted Prime")
else:
    print("Not Twisted Prime")
```

```python
#Emirp Number

n = int(input())

rev = int(str(n)[::-1])

count1 = 0
count2 = 0

for i in range(1, n + 1):

    if n % i == 0:
        count1 += 1

for i in range(1, rev + 1):

    if rev % i == 0:
        count2 += 1

if count1 == 2 and count2 == 2 and n != rev:
    print("Emirp Number")
else:
    print("Not Emirp Number")
```

```python
#max_gap

n = int(input())

b = bin(n)[2:]

max_gap = 0
count = 0

for i in b:

    if i == '0':
        count += 1

    else:
        max_gap = max(max_gap, count)
        count = 0

print(max_gap)
```

```python
#Magic Number

n = int(input())

while n > 9:

    sum = 0

    while n > 0:
        sum += n % 10
        n //= 10

    n = sum

if n == 1:
    print("Magic Number")
else:
    print("Not Magic Number")
```

```python
#Circular Prime
n = input()

flag = True

for i in range(len(n)):

    num = int(n[i:] + n[:i])

    count = 0

    for j in range(1, num + 1):

        if num % j == 0:
            count += 1

    if count != 2:
        flag = False

if flag:
    print("Circular Prime")
else:
    print("Not Circular Prime")
```

```python
#Scenario 1: ATM Card Verification (Binary Search)
arr = [1001, 1002, 1003, 1004, 1005]
target = 1003

low = 0
high = len(arr) - 1

while low <= high:

    mid = (low + high) // 2

    if arr[mid] == target:
        print("Card Valid")
        break

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1
```

```python
#Student Ranking (Selection Sort)
arr = [88, 72, 95, 60, 81]

for i in range(len(arr)):
    min_index = i

    for j in range(i+1, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print(arr)
```

```python
#Card Sorting (Insertion Sort)
arr = [7, 3, 9, 1, 5]

for i in range(1, len(arr)):

    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1

    arr[j+1] = key

print(arr)
```

```python
#Armstrong Number
n = int(input())

s = 0

for i in str(n):
    s += int(i) ** len(str(n))

if s == n:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

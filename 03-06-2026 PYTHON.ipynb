```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_person(self):
        print(self.name, self.age)

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

    def display_student(self):
        print(self.roll_no)

s = Student("Nitya", 20, 101)
s.display_person()
s.display_student()
```

```python
class Shape:
    pass

class Circle(Shape):
    def draw(self):
        print("Circle")

class Rectangle(Shape):
    def draw(self):
        print("Rectangle")

class Triangle(Shape):
    def draw(self):
        print("Triangle")

c = Circle()
r = Rectangle()
t = Triangle()

c.draw()
r.draw()
t.draw()
```

```python
class A:
    def methodA(self):
        print("A")

class B(A):
    def methodB(self):
        print("B")

class C(A):
    def methodC(self):
        print("C")

class D(B, C):
    def methodD(self):
        print("D")

d = D()

d.methodA()
d.methodB()
d.methodC()
d.methodD()
```

```python
class Employee:
    def emp(self):
        print("Employee")

class Manager(Employee):
    def manager(self):
        print("Manager")

class Developer(Employee):
    def developer(self):
        print("Developer")

class TeamLead(Manager, Developer):
    def lead(self):
        print("Team Lead")

t = TeamLead()

t.emp()
t.manager()
t.developer()
t.lead()
```

```python
class Person:

    def __init__(self, name):
        self.name = name


class Student(Person):

    def __init__(self, name, rollno):
        super().__init__(name)
        self.rollno = rollno


s = Student("Nityasri", 101)

print("Name:", s.name)
print("Roll No:", s.rollno)
```

```python
class Vehicle:

    def __init__(self, brand):
        self.brand = brand


class Car(Vehicle):

    def __init__(self, brand, model, price):
        super().__init__(brand)
        self.model = model
        self.price = price


c = Car("Hyundai", "i20", 900000)

print("Brand:", c.brand)
print("Model:", c.model)
print("Price:", c.price)
```

```python
class Account:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance


class SavingsAccount(Account):

    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate


s = SavingsAccount(12345, 50000, 5)

print("Account Number:", s.account_number)
print("Balance:", s.balance)
print("Interest Rate:", s.interest_rate)
```

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age


class Teacher(Person):

    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject


t = Teacher("Ravi", 35, "Python")

print("Name:", t.name)
print("Age:", t.age)
print("Subject:", t.subject)
```

```python
class Product:

    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price


class Electronics(Product):

    def __init__(self, product_name, price, warranty):
        super().__init__(product_name, price)
        self.warranty = warranty


e = Electronics("Laptop", 60000, "2 Years")

print("Product Name:", e.product_name)
print("Price:", e.price)
print("Warranty:", e.warranty)
```
